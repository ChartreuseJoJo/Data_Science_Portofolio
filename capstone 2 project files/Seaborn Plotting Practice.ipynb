# Customizing with matplotlib

```python
df.columns
```

```python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot the distribution of data
sns.histplot(df['AVGFACSAL'], ax=ax)

# Create a more descriptive x axis label
ax.set(xlabel="Average Faculty Salary")

# Show the plot
plt.show()
```

```python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot the distribution of 1 bedroom rents
sns.histplot(df['AVGFACSAL'], ax=ax)

# Modify the properties of the plot
ax.set(xlabel="High End",
       xlim=(5000,15000),
       title="Salaries")

# Display the plot
plt.show()
```

```python
# Create a figure and axes. Then plot the data
fig, ax = plt.subplots()
sns.histplot(df['Tuition'], ax=ax)

# Customize the labels and limits
ax.set(xlabel="1 Bedroom Fair Market Rent", xlim=(10000,50000), title="US Rent")

# Add vertical lines for the median and mean
ax.axvline(x=df['Tuition'].median(), color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=df['Tuition'].mean(), color='b', label='Mean', linestyle='-', linewidth=2)

# Show the legend and plot the data
ax.legend()
plt.show()
```

```python
# Create a plot with 1 row and 2 columns that share the y axis label
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True)

# Plot the distribution of 1 bedroom apartments on ax0
sns.histplot(df['Tuition'], ax=ax0)
ax0.set(xlabel="Tuition")

# Plot the distribution of 2 bedroom apartments on ax1
sns.histplot(df['Tuition'], ax=ax1)
ax1.set(xlabel="2 Bedroom Fair Market Rent", xlim=(1000,15000))

# Display the plot
plt.show()
```

# Joint Grid

```python
# Build a JointGrid comparing humidity and total_rentals
sns.set_style("whitegrid")
g = sns.JointGrid(x="Tuition",
            y="AVGFACSAL",
            data=df,
            xlim=(0,100000)) 

g.plot(sns.regplot, sns.histplot)

plt.show()
plt.clf()
```

```python
# Create a jointplot similar to the JointGrid 
sns.jointplot(x="Tuition",
            y="AVGFACSAL",
        kind='reg',
        data=df)

plt.show()
plt.clf()
```

```python
# Plot temp vs. total_rentals as a regression plot
sns.jointplot(x="Tuition",
            y="AVGFACSAL",
         kind='reg',
         data=df,
         order=1,
         xlim=(0, 60000))

plt.show()
plt.clf()
```

```python
# Plot a jointplot showing the residuals
sns.jointplot(x="Tuition",
            y="AVGFACSAL",
        kind='resid',
        data=df,
        order=2)

plt.show()
plt.clf()
```

```python
# Create a jointplot of temp vs. casual riders
# Include a kdeplot over the scatter plot
g = (sns.jointplot(x="Tuition",
            y="AVGFACSAL",
             kind='scatter',
             data=df,
             marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))
    
plt.show()
plt.clf()
```

```python
# Replicate the above plot but only for registered riders
g = (sns.jointplot(x="Tuition",
            y="AVGFACSAL",
             kind='reg',
             data=df,
             marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))

plt.show()
plt.clf()
```
